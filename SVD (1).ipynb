```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# 이미지 로드 및 회색조 변환
image_path = 'picture.jpg'  # 이미지 파일 경로를 지정
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# 이미지를 numpy 배열로 변환
image_matrix = np.array(image, dtype=float)

# SVD 수행
U, S, VT = np.linalg.svd(image_matrix)

# 특잇값과 특이벡터 분석


# SVD를 사용하여 이미지 복원하기
# 특잇값의 일부만 사용하여 저차원 근사 이미지를 생성
k = 10  # 사용할 특잇값의 수
reconstructed_image = np.dot(U[:, :k], np.dot(np.diag(S[:k]), VT[:k, :]))
reconstructed_image2 = np.dot(U[:, :5*k], np.dot(np.diag(S[:5*k]), VT[:5*k, :]))
reconstructed_image3 = np.dot(U[:, :10*k], np.dot(np.diag(S[:10*k]), VT[:10*k, :]))

# 원본 이미지와 복원된 이미지 시각화
plt.figure(figsize=(10, 5))

plt.subplot(2, 2, 1)
plt.title("original image")
plt.imshow(image, cmap='gray')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.title(f"SVD image (k={k})")
plt.imshow(reconstructed_image, cmap='gray')
plt.axis('off')

plt.subplot(2, 2, 3)
plt.title(f"SVD image (k={5*k})")
plt.imshow(reconstructed_image2, cmap='gray')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.title(f"SVD image (k={10*k})")
plt.imshow(reconstructed_image3, cmap='gray')
plt.axis('off')


plt.show()
```
